## Was ist Redis?
Redis ist eine In-Memory NoSQL-Datenbank, die Daten primär im RAM speichert.  
Sie folgt einem Key-Value-Modell, unterstützt aber auch komplexe Datenstrukturen wie:
- Strings
- Listen
- Sets
- Hashes
- Sorted Sets  

Das Hauptziel von Redis ist es, extrem schnellen Datenzugriff mit sehr geringer Latenz zu ermöglichen.

---

## Was ist das Besondere an Redis?
Redis speichert Daten im Arbeitsspeicher statt auf der Festplatte und ist dadurch extrem schnell.

Es wird häufig als Cache verwendet, kann aber auch als *primäre Datenbank*
mit optionaler Persistenz (AOF, RDB) eingesetzt werden.

Durch Erweiterungen (Module) unterstützt Redis verschiedene Paradigmen und geht damit über ein einfaches Key-Value-System hinaus.

---

## Erweiterte Konzepte
Redis kann durch Module erweitert werden:

- RedisJSON → ermöglicht das Speichern und Abfragen von JSON-Dokumenten  
- RediSearch → ermöglicht Volltextsuche und komplexe Queries  

Damit kann Redis ähnlich wie Dokumentdatenbanken (z. B. MongoDB) verwendet werden,
jedoch mit deutlich höherer Geschwindigkeit durch In-Memory-Speicherung.

---

## Typische Einsatzgebiete
- Caching (z. B. API-Responses, Webseiten)
- Session-Storage (Benutzersitzungen)
- Echtzeitanwendungen (Chat, Gaming, Leaderboards)
- Queues und Messaging-Systeme
- Primäre Datenbank für hochperformante Anwendungen

---

## Stärken
- Sehr hohe Geschwindigkeit (In-Memory)
- Flexible Datenstrukturen
- Kann einen separaten Cache überflüssig machen
- Erweiterbar durch Module (z. B. JSON, Suche)

---

## Schwächen
- Begrenzung durch verfügbaren RAM
- Höhere Kosten im Vergleich zu festplattenbasierten Datenbanken
- Weniger mächtige Query-Funktionalität als vollständige Dokumentdatenbanken
- Persistenz ist optional und nicht standardmäßig aktiv

---

## Dokumentation
- Offizielle Website: https://redis.io  
- Dokumentation: https://redis.io/docs  

---

## Zusammenfassung
Redis ist eine hochperformante In-Memory-Datenbank, die sowohl als Cache
als auch als primäre Datenbank eingesetzt werden kann.

Mit Erweiterungen wie RedisJSON unterstützt sie auch dokumentbasierte Datenmodelle
ähnlich wie MongoDB – bei gleichzeitig deutlich höherer Geschwindigkeit.

## Method 1: Local Installation on Windows 11 with WSL
### 1. Update Packages and Install Redis
sudo apt update
sudo apt install redis-server -y

Downloads and installs the Redis server package.

### 2. Start the Redis Service
sudo service redis-server start

Initializes the Redis background service within the WSL environment. Because WSL does not universally rely on systemd by default, using the service command is the standard method to start daemons in this environment.

### 3. Verify the Connection
redis-cli ping

Accesses the Redis command-line interface and sends a PING command to the local database instance. If the server is active and responding properly, it will return the output PONG.

### 4. Check Service Status
sudo service redis-server status

Outputs the current operational state of the Redis service.

### 5. Stop the Service
sudo service redis-server stop

Halts the active Redis background service.

## 1. Die Basics: Remote Dictionary Server & Commands

Der Name **Redis** steht für **Re**mote **Di**ctionary **S**erver. Das erklärt sofort die Funktionsweise: Es verhält sich wie ein gigantisches Python-Dictionary (Key-Value-Paare), das im Arbeitsspeicher eines Servers liegt.

Anders als relationale Datenbanken (wie MySQL) hat Redis **keine** klassische Abfragesprache wie SQL (`SELECT * FROM table`). Stattdessen arbeitet es mit **Befehlen (Commands)**. Jeder Datentyp hat dabei seine eigenen, hochoptimierten Befehle:

### Das CRUD-Prinzip bei Strings (unser erstes Beispiel)
- Create / Update: `SET` (Erstellt den Key oder überschreibt den alten Wert)
- Read: `GET` (Liest den Wert aus)
- Delete: `DEL` (Löscht den Key)

*Weitere Datentypen, die wir gleich live sehen:*
- **Lists:** `LPUSH`, `RPUSH`, `LRANGE`
- **Hashes:** `HSET`, `HGET`, `HGETALL`

Alle verfügbaren Befehle gibt es hier: [Offizielle Redis Commands](https://redis.io/commands/)

```python
# CREATE
r.set('wichtige_notiz', 'Hallo Kurs! Das ist ein Test.')
```

```python
# READ
print("Aktueller Wert:", r.get('wichtige_notiz'))
```

```python
# UPDATE
r.set('wichtige_notiz', 'UPDATE: Der Text wurde geändert!')
```

```python
# DELETE
r.delete('wichtige_notiz')
```

```python
import time

# Verbindung zu Redis aufbauen
r = redis.Redis(host='localhost', port=6379, decode_responses=True)

# ex=4 setzt ein Ablaufdatum (Time-To-Live) von 4 Sekunden.
r.set('session_token', 'super_geheimes_token_123', ex=4)

print("Wir überwachen das Token für 5 Sekunden:")
for i in range(1, 6):
    wert = r.get('session_token')
    print(f"Sekunde {i}: {wert}")
    time.sleep(1) # Python pausiert für 1 Sekunde
```

## 2. Listen (Arrays) in Redis
Perfekt für Warteschlangen (Queues) oder Protokolle.
- `LPUSH` / `RPUSH`: Elemente vorne oder hinten anfügen.
- `LRANGE`: Liste auslesen.
- `LSET`: Ein Element an einem bestimmten Index updaten!

```python
# Vorherige Daten bereinigen, damit die Liste bei mehrmaligem Ausführen nicht wächst
r.delete('todo_liste')

# RPUSH (Right Push) hängt hinten an die Liste an
r.rpush('todo_liste', 'Jupyter Notebook öffnen')
r.rpush('todo_liste', 'Präsentation halten')

# LPUSH (Left Push) schummelt sich ganz nach vorne!
r.lpush('todo_liste', 'Kaffee trinken') 

# LRANGE holt die Elemente (0 ist der Start, -1 bedeutet "bis zum allerletzten Element")
todos = r.lrange('todo_liste', 0, -1)

print("Meine To-Do-Liste:")
# enumerate generiert uns die Zahlen 1, 2, 3... passend zu den Aufgaben
for index, task in enumerate(todos, 1):
    print(f"{index}. {task}")
```

```python
# Update: Wir ändern das erste Element (Index 0)
r.lset('todo_liste', 0, 'WICHTIG: Viel Kaffee trinken!')
todos = r.lrange('todo_liste', 0, -1)
print("Meine To-Do-Liste:")
for index, task in enumerate(todos, 1):
    print(f"{index}. {task}")
```

## 3. Hashes (Objekte)
Hashes sind quasi "Dictionaries in einem Dictionary". Sie sind ideal, um Objekte zu speichern, die man normalerweise in einer Zeile einer SQL-Tabelle ablegen würde (z.B. Benutzerdaten).

```python
# Wir definieren den Namen unseres Keys als einfachen Python-String
user_key = 'student:1001'

# HSET (Hash Set) speichert ein ganzes Dictionary unter diesem Key ab
r.hset(user_key, mapping={
    'name': 'Max Mustermann',
    'alter': 24,
    'studiengang': 'Wirtschaftsinformatik'
})

# Nur ein einzelnes Attribut (Read) auslesen
name = r.hget(user_key, 'name')
print(f"Nur der Name: {name}\n")

# Ein einzelnes Attribut updaten (das Alter ändern)
r.hset(user_key, 'alter', 25)

print("Das komplette, geupdatete Objekt:")
print(r.hgetall(user_key))
```

## 4. Real-World Szenario: Wie sieht eine "Tabelle" in Redis aus?
Da wir keine Spalten und Zeilen haben, simulieren wir Tabellen oft durch **Schlüssel-Präfixe** (z.B. `student:1`, `student:2`). 

*Aber Achtung:* Wenn wir komplexe Abfragen machen wollen (wie in SQL), stoßen wir mit purem Redis an unsere Grenzen. Hier kommen die Module **RedisJSON** und **RediSearch** ins Spiel!

```python
# Wir erstellen zwei "Zeilen" in unserer Datenbank
r.hset('student:1', mapping={'name': 'Anna', 'note': 1.0})
r.hset('student:2', mapping={'name': 'Ben', 'note': 2.3})

# Wir speichern eine Liste aller bekannten IDs (als Index)
r.delete('alle_studenten')
r.rpush('alle_studenten', 'student:1', 'student:2')

print("--- Unsere simulierte Datenbank-Tabelle ---")
# Wir iterieren über unseren "Index" und holen die Daten
alle_ids = r.lrange('alle_studenten', 0, -1)

for student_id in alle_ids:
    daten = r.hgetall(student_id)
    print(f"ID: {student_id} | Name: {daten['name']} | Note: {daten['note']}")
```

```python
print("\n--- SQL-Suche simuliert in purem Redis ---")
print("Query: SELECT * FROM studenten WHERE note = 1.0\n")

# 1. Wir holen uns die Liste aller bekannten Studenten-IDs
alle_ids = r.lrange('alle_studenten', 0, -1)

# 2. Wir iterieren (loopen) durch alle durch und filtern in Python
for student_id in alle_ids:
    daten = r.hgetall(student_id)
    
    # WICHTIG: Redis speichert (ohne JSON-Modul) alles als Text (String)!
    # Wir müssen '1.0' also als String vergleichen oder in Float umwandeln.
    if float(daten['note']) == 1.0:
        print(f"✅ Treffer gefunden! {daten['name']} hat die Note 1.0")
```

## 4. Das große Finale: RedisJSON & RediSearch in Aktion!

Wie mein Kollege gerade erklärt hat, stoßen wir mit purem Redis an unsere Grenzen, wenn wir komplexe Abfragen machen wollen. Aber mit den Modulen **RedisJSON** und **RediSearch** wird Redis zu einer vollwertigen Dokumentendatenbank!

Wir übersetzen jetzt unsere SQL-Abfrage in Redis:
`SELECT * FROM studenten WHERE note = 1.0`

```python
import json
from redis.commands.search.field import TextField, NumericField

# --- DER ABSOLUT AUSFALLSICHERE IMPORT ---
try:
    # Weg 1: In ganz neuen Versionen direkt aus dem Search-Modul
    from redis.commands.search import IndexDefinition, IndexType
except ImportError:
    try:
        # Weg 2: Der neue Pfad mit Unterstrich (snake_case)
        from redis.commands.search.index_definition import IndexDefinition, IndexType
    except ImportError:
        # Weg 3: Der alte Pfad (mit großem D)
        from redis.commands.search.indexDefinition import IndexDefinition, IndexType
        
r.delete('student:1', 'student:2', 'student:3')        
        
# 1. Wir speichern unsere Daten jetzt als ECHTES JSON (nicht mehr als simplen Hash)
# Das '$' bedeutet: Speichere es an der Wurzel (Root) des JSON-Dokuments
r.json().set('student:1', '$', {'name': 'Anna', 'note': 1.0, 'studiengang': 'WINF'})
r.json().set('student:2', '$', {'name': 'Ben', 'note': 2.3, 'studiengang': 'WINF'})
r.json().set('student:3', '$', {'name': 'Clara', 'note': 1.0, 'studiengang': 'Informatik'})

# 2. Wir erstellen einen Index (Das "Inhaltsverzeichnis" für unsere Suche)
# Wir sagen Redis: "Behalte alle Keys im Auge, die mit 'student:' anfangen, 
# und mach die Felder 'name' (als Text) und 'note' (als Zahl) durchsuchbar."
schema = (
    TextField("$.name", as_name="name"),
    NumericField("$.note", as_name="note")
)
index_def = IndexDefinition(prefix=["student:"], index_type=IndexType.JSON)

try:
    r.ft('idx:studenten').create_index(schema, definition=index_def)
    print("Such-Index erfolgreich erstellt!\n")
except Exception as e:
    # Ignorieren, falls wir die Zelle zweimal ausführen und der Index schon da ist
    pass 

# 3. DIE MAGIE: Unsere SQL-Abfrage in Redis!
# Die Syntax für Zahlen in RediSearch ist @feld:[min max]. 
# Für exakt 1.0 suchen wir also von 1.0 bis 1.0
query = '@note:[1.0 1.0]'
ergebnis = r.ft('idx:studenten').search(query)

print(f"--- ERGEBNIS FÜR: SELECT * FROM studenten WHERE note = 1.0 ---")
print(f"Wir haben {ergebnis.total} brillante Studenten gefunden!\n")

# Wir loopen durch die Ergebnisse
for doc in ergebnis.docs:
    # doc.json ist ein String, wir wandeln ihn kurz um, damit er schöner aussieht
    daten = json.loads(doc.json)
    print(f"ID: {doc.id} | Name: {daten['name']} | Studiengang: {daten['studiengang']}")
```
